# Carregando a base de dados

**a base de dados deve possuir "i#" cado queira ignorar o atributo e "c#" para  a classe**

```python
base_credit = Orange.data.Table('credit_data_regras.csv')
```

```python
base_credit
```

```python
base_credit.domain
```

# Separando Base de dados em Treinamento/Test

```python
base_dividida = Orange.evaluation.testing.sample(base_credit, n = 0.25)
```

```python
base_dividida[0],base_dividida[1]
```

```python
base_treinamento = base_dividida[1]
base_teste = base_dividida[0]
```

```python
len(base_treinamento),len(base_teste)
```

# Aplicando o Algoritmo

```python
cn2 = Orange.classification.rules.CN2Learner()
regras_credit = cn2(base_treinamento)
```

```python
#para prever manualmente
#predict = regras_credit([[a,b,c],[d,e,f]])
#predict
```

```python
for regra in regras_credit.rule_list:
  print(regra)
```

```python
previsoes = Orange.evaluation.testing.TestOnTestData(base_treinamento,base_teste,[lambda testdata:regras_credit])
```

```python
previsoes
```

# Acurácia

```python
Orange.evaluation.CA(previsoes)
```
